```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
bank = pd.read_csv('https://goo.gl/npYFvD')
authentic = bank['authentic'].copy()
bank.drop(labels = ['authentic'], inplace = True, axis = 1)
model = RandomForestClassifier( oob_score = True, random_state = 2)
model.fit(bank, authentic)
print(cross_val_score(model, bank, authentic, cv = 3))
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
df = pd.read_csv('https://goo.gl/MKIGg2')
label = df['class'].copy()
df.drop(labels = 'class', inplace = True, axis = 1)
df_train, df_test, label_train, label_test = train_test_split(df, label,
                                                              test_size = 0.25, 
                                                              random_state = 2)
svc = SVC(C = 0.5)
svc.fit(df_train, label_train)
print(confusion_matrix(label_test, svc.predict(df_test)))
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
wines = pd.read_csv('https://goo.gl/mY06vG', index_col = 0)
labels = wines['quality'].copy()
wines.drop(labels = ['quality'], inplace = True, axis = 1)
parameters = {'C':[1, 3, 5]}
model = SVC()
sv_grid_classifier = GridSearchCV(model, parameters)

sv_grid_classifier.fit(wines, labels)
print(sv_grid_classifier.best_score_)
```

```python
import pandas as pd
from sklearn.manifold import Isomap
movies = pd.read_csv('https://goo.gl/J5JtpN')
iso = Isomap(n_neighbors = 5, n_components = 5)
iso.fit(movies)
transformed_samples = iso.transform(movies.iloc[-3:,:])
print(transformed_samples)
```
